```python
import cv2

INTERP_MAP = {
    "AREA"    : cv2.INTER_AREA,
    "CUBIC"   : cv2.INTER_CUBIC,
    "LANCZOS4": cv2.INTER_LANCZOS4,
}

def to_rgb(img_bgr):
    return cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

def to_bgr(img_rgb):
    return cv2.cvtColor(img_rgb, cv2.COLOR_RGB2BGR)

def read_rgb(path: str):
    img = cv2.imread(str(path), cv2.IMREAD_COLOR)
    if img is None:
        raise FileNotFoundError(f"cv2.imread failed: {path}")
    return to_rgb(img)

def crop_to_scale(img, scale=UPSCALE):
    h, w = img.shape[:2]
    return img[:h - (h % scale), :w - (w % scale), :]

def resize_to_height(img, target_h=TARGET_H, kernel=cv2.INTER_AREA):
    h, w = img.shape[:2]
    if h == target_h:
        return img
    scale = target_h / float(h)
    return cv2.resize(img, (int(round(w*scale)), target_h), interpolation=kernel)

def make_lr_from_hr(hr_rgb, scale=UPSCALE, kernel_name=DOWNSCALE_KERNEL):
    hr_rgb = crop_to_scale(hr_rgb, scale)
    k = INTERP_MAP[kernel_name]
    h, w = hr_rgb.shape[:2]
    lr = cv2.resize(hr_rgb, (w//scale, h//scale), interpolation=k)
    return lr

def chw_from_hwc(img_rgb_uint8):
    return torch.from_numpy(img_rgb_uint8.transpose(2,0,1)).float() / 255.0

def hwc_from_chw(t):
    t = t.clamp(0,1).detach().cpu().numpy().transpose(1,2,0)
    return (t * 255.0 + 0.5).astype(np.uint8)
```

```python
# recursive dataset + fixed-size training patches
from torch.utils.data import Dataset, DataLoader
import glob, numpy as np, torch, os, cv2
from pathlib import Path

def random_jpeg(img_rgb, p=0.3, quality_min=60, quality_max=95):
    # simulate compression artifacts (works on uint8 RGB)
    if np.random.rand() > p:
        return img_rgb
    q = np.random.randint(quality_min, quality_max+1)
    encode_param = [int(cv2.IMWRITE_JPEG_QUALITY), q]
    ok, enc = cv2.imencode('.jpg', cv2.cvtColor(img_rgb, cv2.COLOR_RGB2BGR), encode_param)
    if not ok:
        return img_rgb
    bgr = cv2.imdecode(enc, cv2.IMREAD_COLOR)
    return cv2.cvtColor(bgr, cv2.COLOR_BGR2RGB)

PATCH_HR = 256  # was 192

class SRDataset(Dataset):
    def __init__(self, hr_dir: Path, target_h=TARGET_H, scale=UPSCALE, kernel_name=DOWNSCALE_KERNEL,
                 augment=True, use_patches=True, patch_hr=PATCH_HR, recursive=True):
        exts = (".png",".jpg",".jpeg",".bmp",".tif",".tiff")
        pattern = str(Path(hr_dir)/"**/*") if recursive else str(Path(hr_dir)/"*")
        self.paths = sorted([p for p in glob.glob(pattern, recursive=recursive)
                             if os.path.isfile(p) and p.lower().endswith(exts)])
        if not self.paths:
            raise RuntimeError(f"No images found in {hr_dir}.")
        self.target_h   = target_h
        self.scale      = scale
        self.kernel_name= kernel_name
        self.augment    = augment
        self.use_patches= use_patches
        # ensure patch size is multiple of scale
        self.patch_hr   = int(patch_hr // scale * scale) if use_patches else None

    def __len__(self):
        return len(self.paths)

    def _augment(self, img):
        if np.random.rand() < 0.5: img = np.flip(img, axis=1).copy()  # H flip
        if np.random.rand() < 0.5: img = np.flip(img, axis=0).copy()  # V flip
        k = np.random.randint(0,4)
        if k: img = np.rot90(img, k).copy()
        return img

    def _random_hr_crop(self, hr):
        """Random HR crop of size patch_hr x patch_hr (keeps multiple-of-scale)."""
        ph = self.patch_hr
        H, W = hr.shape[:2]
        if H < ph or W < ph:
            # If image is smaller, center-crop to min available multiple-of-scale
            h = (H // self.scale) * self.scale
            w = (W // self.scale) * self.scale
            y0 = max((H - h)//2, 0); x0 = max((W - w)//2, 0)
            return hr[y0:y0+h, x0:x0+w, :]
        y0 = np.random.randint(0, H - ph + 1)
        x0 = np.random.randint(0, W - ph + 1)
        return hr[y0:y0+ph, x0:x0+ph, :]

    def __getitem__(self, idx):
        # 1) Read & normalize HR (2K height, multiple-of-scale)
        hr = read_rgb(self.paths[idx])
        hr = resize_to_height(hr, target_h=self.target_h, kernel=cv2.INTER_AREA)
        hr = crop_to_scale(hr, self.scale)

        # 2) Optional augmentation
        if self.augment:
            hr = self._augment(hr)
            hr = random_jpeg(hr, p=0.3)

        # 3) Train: random fixed-size patch; Val/Test: use full image
        if self.use_patches and self.patch_hr:
            hr = self._random_hr_crop(hr)

        # 4) Make LR from HR (consistent kernel)
        lr = make_lr_from_hr(hr, scale=self.scale, kernel_name=self.kernel_name)

        # 5) To tensors
        hr_t = chw_from_hwc(hr)
        lr_t = chw_from_hwc(lr)
        return lr_t, hr_t

# Build loaders:
#  - Train uses patches (uniform size -> can batch >1)
#  - Val uses full images (batch_size=1 is fine)
train_loader = DataLoader(
    SRDataset(Path("/content/dataset4k"),
              target_h=TARGET_H, scale=UPSCALE, kernel_name=DOWNSCALE_KERNEL,
              augment=True, use_patches=True, patch_hr=PATCH_HR, recursive=True),
    batch_size=BATCH_SIZE, shuffle=True, num_workers=2, pin_memory=True, drop_last=True
)

val_loader = DataLoader(
    SRDataset(Path("/content/dataset4k_val"),
              target_h=TARGET_H, scale=UPSCALE, kernel_name=DOWNSCALE_KERNEL,
              augment=False, use_patches=False, recursive=False),  # full images
    batch_size=1, shuffle=False, num_workers=2, pin_memory=True
)

print("Batches — train:", len(train_loader), "| val:", len(val_loader))
```

```python
import torch.nn as nn

class ResidualBlockBN(nn.Module):
    def __init__(self, c=64):
        super().__init__()
        self.conv1 = nn.Conv2d(c, c, 3, 1, 1)
        self.bn1 = nn.BatchNorm2d(c)
        self.relu = nn.ReLU(inplace=True)
        self.conv2 = nn.Conv2d(c, c, 3, 1, 1)
        self.bn2 = nn.BatchNorm2d(c)

    def forward(self, x):
        y = self.conv1(x); y = self.bn1(y); y = self.relu(y)
        y = self.conv2(y); y = self.bn2(y)
        return x + y

class SuperResolutionNet(nn.Module):
    def __init__(self, num_blocks=8, channels=64, scale=4):
        super().__init__()
        self.head = nn.Conv2d(3, channels, 3, 1, 1)
        self.body = nn.Sequential(*[ResidualBlockBN(channels) for _ in range(num_blocks)])
        up = []
        s = scale
        while s > 1:
            up.append(nn.Conv2d(channels, channels*4, 3, 1, 1))
            up.append(nn.PixelShuffle(2))
            up.append(nn.ReLU(inplace=True))
            s //= 2
        self.upsample = nn.Sequential(*[m for m in up])
        self.tail = nn.Conv2d(channels, 3, 3, 1, 1)

    def forward(self, x):
        x = self.head(x)
        x = self.body(x)
        x = self.upsample(x)
        x = self.tail(x)
        return x

model = SuperResolutionNet(scale=UPSCALE).to(DEVICE)
total_params = sum(p.numel() for p in model.parameters())
print(f"Model params: {total_params/1e6:.2f} M")
```

```python
import time
import torch.optim as optim
import torch.nn.functional as F
from tqdm import tqdm
import numpy as np

optimizer = optim.Adam(model.parameters(), lr=LR)
scaler = torch.cuda.amp.GradScaler(enabled=(DEVICE.type=="cuda"))

def psnr_from_tensors(sr, hr, eps=1e-12):
    # Expect [0,1] floats; compute per-batch MSE then PSNR
    mse = F.mse_loss(sr, hr, reduction="mean").item()
    return 10.0 * np.log10(1.0 / (mse + eps))

@torch.no_grad()
def eval_psnr(model, loader):
    model.eval()
    total, n = 0.0, 0
    for lr, hr in loader:
        lr, hr = lr.to(DEVICE), hr.to(DEVICE)
        with torch.cuda.amp.autocast(enabled=(DEVICE.type=="cuda")):
            sr = model(lr).clamp(0,1)
        total += psnr_from_tensors(sr, hr); n += 1
    return total / max(n,1)

best_psnr = -1e9
for epoch in range(1, EPOCHS+1):
    model.train()
    t0 = time.time()
    running = 0.0
    for lr, hr in tqdm(train_loader, desc=f"Epoch {epoch}/{EPOCHS}", leave=False):
        lr, hr = lr.to(DEVICE), hr.to(DEVICE)
        optimizer.zero_grad(set_to_none=True)
        with torch.cuda.amp.autocast(enabled=(DEVICE.type=="cuda")):
            sr = model(lr)
            loss = F.mse_loss(sr, hr)
        scaler.scale(loss).backward()
        scaler.step(optimizer)
        scaler.update()
        running += loss.item()

    val_psnr = eval_psnr(model, val_loader)
    dt = time.time() - t0
    print(f"[{epoch:03d}/{EPOCHS}] train_loss={running/len(train_loader):.5f}  val_PSNR={val_psnr:.2f} dB  time={dt:.1f}s")

    if val_psnr > best_psnr:
        best_psnr = val_psnr
        torch.save(model.state_dict(), BEST_PATH)
        print(f"  ↳ saved best: {BEST_PATH} (val_PSNR={best_psnr:.2f} dB)")

torch.save(model.state_dict(), WEIGHTS_PATH)
print("Saved final weights:", WEIGHTS_PATH)
```

```python
import csv, numpy as np, torch, cv2
from skimage.metrics import structural_similarity as ssim

@torch.no_grad()
def eval_bicubic_vs_model(model, loader, out_csv="/content/sr_eval_results.csv", max_images=None):
    model.eval()
    rows = []
    for idx, (lr, hr) in enumerate(loader):
        if max_images is not None and idx >= max_images:
            break

        lr, hr = lr.to(DEVICE), hr.to(DEVICE)

        # Bicubic upscale to HR size
        n, c, h_lr, w_lr = lr.shape
        H, W = h_lr*UPSCALE, w_lr*UPSCALE
        lr_img = hwc_from_chw(lr[0])
        bic_img = cv2.resize(lr_img, (W, H), interpolation=cv2.INTER_CUBIC)

        # Model prediction
        with torch.amp.autocast('cuda', enabled=(DEVICE.type=="cuda")):
            sr = model(lr).clamp(0,1)
        hr_img = hwc_from_chw(hr[0])
        sr_img = hwc_from_chw(sr[0])

        # Align shapes defensively
        Hh, Ww = hr_img.shape[:2]
        sr_img  = sr_img[:Hh,:Ww,:]
        bic_img = bic_img[:Hh,:Ww,:]

        # PSNR (on [0,1] tensors)
        def to_float_tensor(img_uint8):
            return torch.from_numpy(img_uint8.transpose(2,0,1)).unsqueeze(0).float() / 255.0
        psnr_m = psnr_from_tensors(to_float_tensor(sr_img), to_float_tensor(hr_img))
        psnr_b = psnr_from_tensors(to_float_tensor(bic_img), to_float_tensor(hr_img))

        # SSIM (on [0,1] floats)
        hr_f  = hr_img.astype(np.float32)/255.0
        sr_f  = sr_img.astype(np.float32)/255.0
        bic_f = bic_img.astype(np.float32)/255.0
        ssim_m = ssim(hr_f, sr_f,  channel_axis=2, data_range=1.0)
        ssim_b = ssim(hr_f, bic_f, channel_axis=2, data_range=1.0)

        rows.append({
            "idx": idx,
            "psnr_model": round(psnr_m, 4),
            "psnr_bicubic": round(psnr_b, 4),
            "delta_psnr": round(psnr_m - psnr_b, 4),
            "ssim_model": round(ssim_m, 5),
            "ssim_bicubic": round(ssim_b, 5),
            "delta_ssim": round(ssim_m - ssim_b, 5),
        })

    # Write CSV (guard for empty just in case)
    if rows:
        with open(out_csv, "w", newline="") as f:
            writer = csv.DictWriter(f, fieldnames=list(rows[0].keys()))
            writer.writeheader(); writer.writerows(rows)

        dpsnr = np.mean([r["delta_psnr"] for r in rows])
        dssim = np.mean([r["delta_ssim"] for r in rows])
        pm = np.mean([r["psnr_model"] for r in rows]); pb = np.mean([r["psnr_bicubic"] for r in rows])
        sm = np.mean([r["ssim_model"] for r in rows]); sb = np.mean([r["ssim_bicubic"] for r in rows])

        print(f"Eval images: {len(rows)}")
        print(f"Model PSNR: {pm:.3f} dB | Bicubic: {pb:.3f} dB | ΔPSNR: {dpsnr:+.3f} dB")
        print(f"Model SSIM: {sm:.4f}   | Bicubic: {sb:.4f}   | ΔSSIM: {dssim:+.4f}")
        print(f"Saved per-image results to: {out_csv}")
    else:
        print("No images evaluated.")

# Load best and run eval on ONE image
state = torch.load(str(BEST_PATH), map_location=DEVICE)
model.load_state_dict(state); model.eval()
eval_bicubic_vs_model(model, val_loader, out_csv="/content/sr_eval_results_head.csv", max_images=1)
```

```python
import math

@torch.no_grad()
def infer_tiled(model, img_rgb_uint8, scale=UPSCALE, tile=512, overlap=32):
    model.eval()
    img = img_rgb_uint8.astype(np.float32) / 255.0
    h, w = img.shape[:2]
    tiles_x = math.ceil(w / tile)
    tiles_y = math.ceil(h / tile)
    out = np.zeros((h*scale, w*scale, 3), dtype=np.float32)
    weight = np.zeros_like(out)

    for ty in range(tiles_y):
        for tx in range(tiles_x):
            x0 = max(tx*tile - overlap, 0)
            y0 = max(ty*tile - overlap, 0)
            x1 = min((tx+1)*tile + overlap, w)
            y1 = min((ty+1)*tile + overlap, h)

            tile_rgb = img[y0:y1, x0:x1, :]
            lr_t = torch.from_numpy(tile_rgb.transpose(2,0,1)).unsqueeze(0).float().to(DEVICE)
            with torch.cuda.amp.autocast(enabled=(DEVICE.type=="cuda")):
                sr_t = model(lr_t).clamp(0,1)
            sr_img = sr_t.squeeze(0).cpu().numpy().transpose(1,2,0)

            sy0, sx0 = y0*scale, x0*scale
            sy1, sx1 = y1*scale, x1*scale

            # Build feathering mask to blend overlaps
            wy = np.linspace(0, 1, sr_img.shape[0], dtype=np.float32)
            wx = np.linspace(0, 1, sr_img.shape[1], dtype=np.float32)
            mask = np.minimum.outer(np.minimum(wy, wy[::-1]), np.minimum(wx, wx[::-1]))
            mask = np.power(mask, 0.5)[..., None]  # soft edges

            out[sy0:sy1, sx0:sx1, :] += sr_img * mask
            weight[sy0:sy1, sx0:sx1, :] += mask

    out = np.divide(out, np.maximum(weight, 1e-8))
    out = np.clip(out, 0, 1)
    return (out * 255.0 + 0.5).astype(np.uint8)
```

```python
# ==== Visualize LR / Bicubic / SR / HR (inline) ====
import os, cv2, numpy as np, matplotlib.pyplot as plt

def _rgb(path):
    bgr = cv2.imread(path, cv2.IMREAD_COLOR)
    if bgr is None:
        raise FileNotFoundError(path)
    return cv2.cvtColor(bgr, cv2.COLOR_BGR2RGB)

# If we still have lr/hr/bic/sr variables from the previous cell, use them.
# Otherwise load from saved files.
def _get_or_load():
    paths = {
        "LR": "/content/demo_lr.png",
        "HR": "/content/demo_hr.png",
        "Bicubic": "/content/demo_bic.png",
        "SR": "/content/demo_sr.png",
    }
    imgs = {}
    try:
        # try the in-memory variables first
        imgs["LR"] = lr if isinstance(lr, np.ndarray) else _rgb(paths["LR"])
        imgs["HR"] = hr if isinstance(hr, np.ndarray) else _rgb(paths["HR"])
        imgs["Bicubic"] = bic if isinstance(bic, np.ndarray) else _rgb(paths["Bicubic"])
        imgs["SR"] = sr if isinstance(sr, np.ndarray) else _rgb(paths["SR"])
    except NameError:
        # fall back to disk
        for k,p in paths.items():
            imgs[k] = _rgb(p)
    return imgs

def show_main_and_zoom(imgs, zoom_size=200):
    # Main row
    titles = ["LR (x4 input)", "Bicubic ↑", "SR (model)", "HR (target)"]
    arr = [imgs["LR"], imgs["Bicubic"], imgs["SR"], imgs["HR"]]

    plt.figure(figsize=(18,4))
    for i, (im, t) in enumerate(zip(arr, titles), 1):
        plt.subplot(1,4,i)
        plt.imshow(im)
        plt.title(t)
        plt.axis("off")
    plt.tight_layout()
    plt.show()

    # Zoomed crop from image center (same region across all)
    H, W = imgs["HR"].shape[:2]
    z = zoom_size
    cy, cx = H//2, W//2
    y0, y1 = max(cy - z//2, 0), min(cy + z//2, H)
    x0, x1 = max(cx - z//2, 0), min(cx + z//2, W)

    # make sure all are same shape as HR; they should be after the demo cell
    Z = {
        "Bicubic": imgs["Bicubic"][y0:y1, x0:x1],
        "SR":      imgs["SR"][y0:y1, x0:x1],
        "HR":      imgs["HR"][y0:y1, x0:x1],
    }

    plt.figure(figsize=(12,4))
    for i, (k) in enumerate(["Bicubic","SR","HR"], 1):
        plt.subplot(1,3,i)
        plt.imshow(Z[k])
        plt.title(f"{k} (center crop)")
        plt.axis("off")
    plt.tight_layout()
    plt.show()

imgs = _get_or_load()
show_main_and_zoom(imgs, zoom_size=220)
```
